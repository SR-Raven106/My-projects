```python
import time

# Vitamin dosage calculation function
def calculate_vitamin_dosage(height, weight, age):
    dosage = (weight / height) * 10  # Example formula for demonstration
    return round(dosage, 2)

# Symptom analysis function using built-in split
def analyze_symptoms(symptoms_input):
    keywords = ['cough', 'fever', 'fatigue', 'headache', 'nausea']
    symptoms_found = [word for word in symptoms_input.lower().split() if word in keywords]
    return symptoms_found

# Reminder function
def send_reminder(message, delay=5):
    print("Sending reminder...")
    time.sleep(delay)  # Simulate a delay (e.g., 5 seconds)
    print(f"Reminder: {message}")

# Example usage
# Input data for vitamin dosage calculation
height = 170  # in cm
weight = 70   # in kg
age = 30      # in years

# Calculate and display vitamin dosage
dosage = calculate_vitamin_dosage(height, weight, age)
print(f"Recommended vitamin dosage: {dosage} units")

# Input data for symptom analysis
symptoms_input = "I have a fever and a headache, but no cough or fatigue."
matched_symptoms = analyze_symptoms(symptoms_input)

# Display matched symptoms
if matched_symptoms:
    print(f"Matched symptoms: {matched_symptoms}")
else:
    print("No known symptoms found.")

# Send a simulated reminder
send_reminder("Don't forget to take your vitamins at 8 AM!", delay=3)

```

```python
from sqlalchemy import create_engine, Column, Integer, String, ForeignKey, UniqueConstraint, Text
from sqlalchemy.orm import declarative_base, relationship, sessionmaker
from getpass import getpass

Base = declarative_base()

class User(Base):
    __tablename__ = 'users'
    id = Column(Integer, primary_key=True, autoincrement=True)
    username = Column(String, unique=True, nullable=False)
    password = Column(String, nullable=False)
    role = Column(String, nullable=False)

    __table_args__ = (UniqueConstraint('username', name='_username_uc'),)

class Message(Base):
    __tablename__ = 'messages'
    id = Column(Integer, primary_key=True, autoincrement=True)
    sender_id = Column(Integer, ForeignKey('users.id'), nullable=False)
    receiver_id = Column(Integer, ForeignKey('users.id'), nullable=False)
    content = Column(Text, nullable=False)

    sender = relationship("User", foreign_keys=[sender_id])
    receiver = relationship("User", foreign_keys=[receiver_id])

class Note(Base):
    __tablename__ = 'notes'
    id = Column(Integer, primary_key=True, autoincrement=True)
    doctor_id = Column(Integer, ForeignKey('users.id'), nullable=False)
    patient_id = Column(Integer, ForeignKey('users.id'), nullable=False)
    content = Column(Text, nullable=False)

    doctor = relationship("User", foreign_keys=[doctor_id])
    patient = relationship("User", foreign_keys=[patient_id])

engine = create_engine('sqlite:///telehealth_app.db')
Base.metadata.create_all(engine)

Session = sessionmaker(bind=engine)
session = Session()

def create_user():
    username = input("Enter username: ")
    password = getpass("Enter password: ")
    role = input("Enter role (patient/doctor): ").lower()

    if role not in ['patient', 'doctor']:
        print("Invalid role. Please choose 'patient' or 'doctor'.")
        return

    new_user = User(username=username, password=password, role=role)
    session.add(new_user)
    session.commit()
    print(f"{role.capitalize()} account for '{username}' created successfully!")

def authenticate_user():
    username = input("Enter username: ")
    password = getpass("Enter password: ")

    user = session.query(User).filter_by(username=username, password=password).first()
    if user:
        print(f"Welcome, {user.username}! You are logged in as a {user.role}.")
        return user
    else:
        print("Invalid username or password.")
        return None

def send_message(sender):
    receiver_username = input("Enter the username of the receiver: ")
    receiver = session.query(User).filter_by(username=receiver_username).first()

    if not receiver:
        print("Receiver not found.")
        return

    if sender.role == 'patient' and receiver.role != 'doctor':
        print("Patients can only send messages to doctors.")
        return

    content = input("Enter your message: ")
    message = Message(sender_id=sender.id, receiver_id=receiver.id, content=content)
    session.add(message)
    session.commit()
    print("Message sent successfully!")

def view_messages(user):
    received_messages = session.query(Message).filter_by(receiver_id=user.id).all()
    print(f"Messages for {user.username}:")
    for msg in received_messages:
        sender = session.query(User).filter_by(id=msg.sender_id).first()
        print(f"From {sender.username}: {msg.content}")

def record_note(doctor):
    if doctor.role != 'doctor':
        print("Only doctors can record notes.")
        return

    patient_username = input("Enter the username of the patient: ")
    patient = session.query(User).filter_by(username=patient_username, role='patient').first()

    if not patient:
        print("Patient not found.")
        return

    content = input("Enter note content: ")
    note = Note(doctor_id=doctor.id, patient_id=patient.id, content=content)
    session.add(note)
    session.commit()
    print("Note recorded successfully!")

while True:
    print("\n1. Create User")
    print("2. Login")
    print("3. Exit")
    choice = input("Choose an option: ")

    if choice == '1':
        create_user()
    elif choice == '2':
        user = authenticate_user()
        if user:
            while True:
                print("\n1. Send Message")
                print("2. View Messages")
                if user.role == 'doctor':
                    print("3. Record Note")
                print("4. Logout")
                user_choice = input("Choose an option: ")

                if user_choice == '1':
                    send_message(user)
                elif user_choice == '2':
                    view_messages(user)
                elif user_choice == '3' and user.role == 'doctor':
                    record_note(user)
                elif user_choice == '4':
                    break
                else:
                    print("Invalid option.")
    elif choice == '3':
        break
    else:
        print("Invalid option.")

existing_user = session.query(User).filter_by(username='BenJ').first()
if not existing_user:
    default_user = User(username='BenJ', password='12345', role='doctor')
    session.add(default_user)
    session.commit()
    print("Predefined user 'BenJ' created with password '12345'.")
else:
    print("Predefined user 'BenJ' already exists.")
```
